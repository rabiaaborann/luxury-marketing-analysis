```python
%load_ext autoreload
%autoreload 2

import sys
import os
from pathlib import Path
import pandas as pd 

# Set project root and append to sys.path
root_dir = Path(os.getcwd()).parent
if str(root_dir) not in sys.path:
    sys.path.append(str(root_dir))

print(f"Project root initialized.")
```

```python
from src.data_fetcher import LuxuryDataFetcher
import time

fetcher = LuxuryDataFetcher()

# Strategic Brand Groups (Group A: High Reach vs Group B: Quiet Luxury)
research_groups = {
    "Group A (High Visibility)": ["miumiu", "ysl", "jacquemus"],
    "Group B (Low Visibility)": ["hermes", "brunellocucinelli_brand", "loropiana"]
}

all_data = []

for group, accounts in research_groups.items():
    for account in accounts:
        url = f"https://www.instagram.com/{account}/"
        print(f"[PROCESS] Fetching 1-year historical data for: {account}...")
        
        # Historical fetch (1 year)
        df = fetcher.fetch_historical_data(url, years=1)
        
        if df is not None:
            df['Brand_Name'] = account
            df['Strategic_Group'] = group
            all_data.append(df)
            
        time.sleep(2) # API safety delay

# Combine into Master Research DataFrame
master_df = pd.concat(all_data, ignore_index=True)
print(f"\n[SUCCESS] Extraction complete. Total posts for analysis: {len(master_df)}")
```

```python
from src.analytics import LuxuryAnalyticsEngine

# Initialize Analytics Engine
engine = LuxuryAnalyticsEngine(master_df)

# Run full paradox analysis
final_report = engine.perform_full_analysis()

# Display results sorted by Stability Score
print("\n--- STRATEGIC LUXURY PARADOX REPORT ---")
display(final_report.sort_values(by="ESS_Score", ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(16, 10), facecolor='#0D0D0D')
ax.set_facecolor('#0D0D0D')

brand_colors = {
    "Group A (High Visibility)": "#D44D5C", 
    "Group B (Low Visibility)": "#B38B4D"
}

# Bubble size is driven by Comment_Quality for added depth
bubble_size = final_report['Comment_Quality'] * 1500

sns.scatterplot(
    data=final_report, 
    x="Volatility_Index", 
    y="ESS_Score",
    hue="Group", 
    size=bubble_size,
    sizes=(400, 2500),
    palette=brand_colors, 
    edgecolors='white', 
    linewidth=0.5, 
    alpha=0.85, 
    ax=ax,
    legend='brief'
)

for i in range(final_report.shape[0]):
    ax.text(
        final_report.Volatility_Index[i], 
        final_report.ESS_Score[i] + 0.2, 
        final_report.Brand[i].upper(),
        color='white', 
        fontsize=10, 
        ha='center',
        alpha=0.9
    )

plt.xlim(0.5, 2.1) # Including Miu Miu's extreme volatility
plt.ylim(0, 6.0)

ax.grid(True, linestyle=':', alpha=0.1, color='#FFFFFF')
sns.despine(left=True, bottom=True)

plt.title("THE EXCLUSIVITY vs. REACH PARADOX", fontsize=22, color="#FFFEFD", pad=30)
plt.xlabel("VOLATILITY INDEX (MARKET SENSITIVITY)", fontsize=11, color='gray', labelpad=15)
plt.ylabel("EXCLUSIVITY STABILITY SCORE (ESS)", fontsize=11, color='gray', labelpad=15)

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False, labelcolor='white')

plt.tight_layout()
plt.show()
```
